```python
import torch
```

Here, I am creating basic Tensors

```python
#Here, I am getting float32 as the datatype as it is the default datatype for PyTorch tensors.
d = torch.zeros((5, 5))
print(d)
print(d.dtype)
print(d.device)
```

```python
#Here, I am changing the datatype to int32 and the device to GPU if available.
d = torch.zeros((5, 5), dtype=torch.int32, device='cuda' if torch.cuda.is_available() else 'cpu')
print(d)
```

```python
a = torch.ones((5, 5), dtype=torch.int16)
print(a)
print(a.dtype)
print(a.device)
```

Here, I am generating random numbers with the help of seed also, generating same number with the help of seed() as it set the starting point of numbers and then sequence follow.

```python
torch.manual_seed(42)
r1 = torch.rand((2,2))
print(r1)

r2 = torch.rand((2, 2))
print('Different random numbers:\n', r2)

torch.manual_seed(42)
r3 = torch.rand((2, 2))
print('Same random numbers as r1:\n', r3)    

```

Here, I am doing arithmcatic opeartions on tensors.

```python
ones = torch.ones((2, 3))

twos = ones * 2 # here it multiplies each element in the tensor by 2

threes = ones + twos # here it adds ones and twos to each element in the tensor
print(threes)
print(threes.dtype)
print(threes.shape)
```

```python
# Here you can not add tensors of different shapes, so the following line will raise an error.
r1 = torch.rand((2, 3))
r2 = torch.rand((3, 2))

print(r1 + r2) # element-wise addition
```

```python
# Example of mathematical opeartions available in Pytorch

# a random metrix of shape (2, 2) with values between -1 and 1
m1 = (torch.rand(2,2)-0.5) * 2
print(m1)

# absolute value
print(torch.abs(m1))

# trignometric functions
print(torch.asin(m1))

# 
```
