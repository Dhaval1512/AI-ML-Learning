```python
import numpy as np

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training data
X = np.array([1, 2, 3, 5, 7]).reshape(-1, 1)
y = np.array([0, 0, 0, 1, 1])

# Initialize parameters
w = 0
b = 0
lr = 0.1

# Training (Gradient Descent)
for i in range(1000):
    z = w * X + b
    p = sigmoid(z)
    
    # gradients
    dw = np.mean((p - y) * X)
    db = np.mean(p - y)
    
    # update
    w -= lr * dw
    b -= lr * db

print("Weight:", w)
print("Bias:", b)

# Prediction
def predict(x):
    prob = sigmoid(w * x + b)
    return 1 if prob >= 0.5 else 0

print("Prediction for 4 hours:", predict(4))
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

X = np.array([1, 2, 3, 5, 7]).reshape(-1, 1)
y = np.array([0, 0, 0, 1, 1])

model = LogisticRegression()
model.fit(X, y)

# Prediction
print(model.predict([[4]]))        # Class
print(model.predict_proba([[4]]))  # Probability
```
